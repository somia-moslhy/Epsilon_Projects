## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to guess the fruit name correctly.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible in the output.
> ##### In this way, all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, the number of chances equals the word length + 2.
> ##### For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.
> ##### Make game repeats until player ends the game.
> ##### Bonus : Create a function that takes 3 parameters `Easy`, `Medium` and `Hard` where `Medium` has more difficult words to be guessed than `Easy` and so on.

```python
import random
```

```python
def choosed_word(level):
    easy=["apple", "banana", "grape"]
    medium=["cherry", "date", "kiwi"]
    hard=["fig", "pomegranate", "watermelon"]

    if level.lower()=="easy":
        return random.choice(easy)
    elif level.lower()=="medium":
        return random.choice(medium)        
    elif level.lower()=="hard":
        return random.choice(hard)
    
def HangmanGame():
    level=input("Choose the difficulty level (easy, medium, hard): ")
    word=choosed_word(level)

    gussed_letters=[]
    chances=len(word)+2 # number of chances
    display_word=["_"]*len(word) # list to display the guessed letters and underscores

    print("Lets, start the game!")
    print("Word to guess: ", " ".join(display_word))
    
    while chances>0:# while there are chances left
        guess=input("Guess a letter: ").lower()
        if len(guess)!=1 or not guess.isalpha(): # only single alphabetical character allowed
            print("Please enter a single alphabetical character.")
            continue

        if guess in gussed_letters:
            print("You already guessed that letter. Try again.")
            continue
        gussed_letters.append(guess)
        if guess in word:
            for i in range(len(word)):
                if word[i]==guess:
                    display_word[i]=guess # replace underscore with correct letter
                    print("Correct")
        else:
            chances-=1
            print("Wrong guess. Chances left:", chances)
        print("Word to guess: ", " ".join(display_word))
        
        if "_" not in display_word:
            print("Congratulations! You've guessed the word:", word)
            break
        
    
    print("Game over! The word was:", word)


```

```python
HangmanGame()
```
